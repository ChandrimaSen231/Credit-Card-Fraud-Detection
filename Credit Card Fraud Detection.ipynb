```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df["Amount"]
```

```python
df["Amount"] = scaler.fit_transform(np.array(df["Amount"]).reshape(-1,1))
```

```python
df["Amount"]
```

```python
df["Amount"].isna().sum()
```

```python
X = df.drop("Class",axis=1)
y = df["Class"]
```

```python
X.head()
```

```python
y
```

## Split the data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.5,random_state=42)
```

```python
X_train.head()
```

```python
y_train.head()
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
clf = DecisionTreeClassifier()

clf = clf.fit(X_train,y_train)
y_train_pred = clf.predict(X_train)
y_pred = clf.predict(X_test)
```

```python
print("Train Data : ")
print("Accuracy : ",accuracy_score(y_train,y_train_pred))
print("Confusion Matrix : ")
print(confusion_matrix(y_train,y_train_pred))
print("Test Data : ")
print("Accuracy : ",accuracy_score(y_test,y_pred))
print("Confusion Matrix : ")
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn import tree
```

```python
import graphviz

dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph
```

# Hyperparemeter Tuning

```python
from sklearn.model_selection import GridSearchCV

params = {
    "criterion" : ["entropy","gini"],
    "splitter" : ["best","random"],
    "max_depth" : [1,2,3,4,5]
}

grid_search = GridSearchCV(clf,params,scoring= 'neg_mean_squared_error')

grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
rank = grid_search.cv_results_["rank_test_score"]
test_score = grid_search.cv_results_["mean_test_score"]
param = grid_search.cv_results_["params"]

for r,t,p in zip(rank,test_score,param):
    error = np.sqrt(-t)
    print(f"{r}. {error:.2f} {p}")
```

#### Evaluating the best Models

```python
clf = DecisionTreeClassifier(criterion='entropy',max_depth=5,splitter='best')

clf = clf.fit(X_train,y_train)
y_train_pred = clf.predict(X_train)
y_pred = clf.predict(X_test)
```

```python
print("Train Data : ")
print("Accuracy : ",accuracy_score(y_train,y_train_pred))
print("Confusion Matrix : ")
print(confusion_matrix(y_train,y_train_pred))
print("Test Data : ")
print("Accuracy : ",accuracy_score(y_test,y_pred))
print("Confusion Matrix : ")
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
